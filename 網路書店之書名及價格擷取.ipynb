```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

books = pd.Series()
prices = pd.Series()

for page in range(1,13):
  time.sleep(15)
  urls = "https://search.books.com.tw/search/query/cat/19/qsub/001/sort/1/v/0/page/" + format(str(page)) + "/ms2/ms2_1/key/python"

  res = requests.get(urls)

  soup = BeautifulSoup(res.text,'html.parser')

  soup.select("tbody > tr > td > h4 > a")
  soup.select("tbody > tr > td > ul.list-nav.clearfix > li")
  
  
  for book in soup.select("tbody > tr > td > h4 > a"):
    books = books.append(pd.Series([book['title']])).reset_index(drop=True)  #加到pd.Series

  i = 0

  
  for price in soup.select("tbody > tr > td > ul.list-nav.clearfix > li"):
    if(i<books.size):
        if(len(price.select('strong'))==1): #只有價格
            prices = prices.append(pd.Series([price.select('strong')[0].string])).reset_index(drop=True) # .string取tag<b>中的文字內容
        elif(len(price.select('strong'))==2): #有打折數+價格
            prices = prices.append(pd.Series([price.select('strong')[1].string])).reset_index(drop=True) # .string取tag<b>中的文字內容
        else:
            break
    i+=1

  df = pd.DataFrame({'書名':books, '價格':prices})
  if page == 1:
    dfs = df
  else:
    dfs = df.append(df,ignore_index = True)
  if page == 12:
    print(df)
```
